```python
# Import useful libraries and classes.
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
#html upload
my_url= "https://www.lenovo.com/in/en/laptops/c/LAPTOPS#new-products?IPromoID=LEN325622"
uClient= uReq(my_url)
page_html= uClient.read()
uClient.close()
```

```python
page_html
```

```python
#html parser
page_soup= soup(page_html)
page_soup
```

```python
#read class from web page.
containers= page_soup.findAll("div", {"class": "new-products-container"})
print(len(containers))
print(containers)
```
